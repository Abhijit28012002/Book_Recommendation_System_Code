```python
# marge book and rating togather
ratings_with_books = ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
ratings_with_books.info()
```

```python
number_ratings = ratings_with_books.groupby('Title')['Book-Rating'].count().reset_index()
```

```python
number_ratings.rename(columns={'Book-Rating':'number_of_rating'},inplace=True)
```

```python
number_ratings
```

```python
final_rating = ratings_with_books.merge(number_ratings,on='Title')
```

```python
final_rating
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['number_of_rating']>=50]
```

```python
final_rating
```

```python
final_rating.shape
```

```python
final_rating.info()
```

```python
final_rating.drop_duplicates(['User-ID','Title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='User-ID',index='Title',values='Book-Rating')
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
print(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances , suggestions = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```
